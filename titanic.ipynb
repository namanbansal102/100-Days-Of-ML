```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('titanic.csv')
```

```python
df.head()
```

```python
sns.barplot(x=df['Pclass'],y=df['Age'],hue=df['Sex'])
```

```python
## Comparing Numberical and
```

```python
sns.boxplot(x=df['Age'],y=df['Sex'],hue=df['Survived'])
```

```python
sns.distplot(df['Age'])
```

```python
sns.distplot(df[df['Survived']==0]['Age'],hist=False)
sns.distplot(df[df['Survived']==1]['Age'],hist=False)
```

```python
# Printing heat Map(Categorical and Categorical)
df.head(3)
```

```python
sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```

```python
(df.groupby('Pclass').mean()['Survived']*100).plot(kind='bar')
```

```python
# ClusterMap(Categorical-CAtegorical)
sns.clustermap(pd.crosstab(df['SibSp'],df['Survived']))
```

```python
# PairPlot
sns.pairplot(df,kind='kde')
```

```python
#LinePlot("Numerical-Numerical")
df.groupby('Age').sum().reset_index()
```
