```python
# Working with csv File Formats
import pandas as pd
```

```python
# Sep Operator
# as sTab Separator file is Separated By Tab
df=pd.read_csv('aug_train.csv',sep=',')
# as Generally Seperated By , Separator
df
```

```python
# Naming The Columns in Pandas DataFrame if names not given In DataFrame
df=pd.read_csv("aug_train.csv",names=['custom_id','zip_code'])
```

```python
# Naming The Index To Enrollment Id
df=pd.read_csv('aug_train.csv',index_col=['enrollee_id'])
```

```python
# using The Row as A Header Parameter
pd.read_csv('aug_train.csv',header=3)
```

```python
# using The 1 Column Id
pd.read_csv('aug_train.csv',usecols=['city'])
```

```python
# using The Squeeze Operator For Getting 1 Column
pd.read_csv('aug_train.csv',usecols=['gender'])
```

```python
# Using The Python Skip Row Parameter To Skip Row in a Pandas DataFrame
# as We Can Write a Function To Skip Rows
# pd.read_csv('aug_train.csv',skiprows=[0,1,2,3,4])
pd.read_csv('aug_train.csv',skiprows=lambda x:x/5==0)
```

```python
# For Skipping The Rows in Pandas DataFrame
# As This is Useful For Getting Number of Rows
pd.read_csv('aug_train.csv',nrows=100)
```

```python
# Encoding Parameter
# As There Are Some Datasets With Different Encodings as Default Encoding is utf-8
pd.read_csv('aug_train.csv',encoding='utf-8')
```

```python
# Skip Bad Lines
pd.read_csv('aug_train.csv',on_bad_lines='skip')
```

```python
# use a Convertor parameter to Change Type of Paramter
def rename(text):
    if text=="Graduate":
        return "UnderGraduate"
    else:
        return "Any Other Course"
pd.read_csv('aug_train.csv',converters={'education_level':rename})
```

```python
# using na Parameter To get Change
pd.read_csv('aug_train.csv',na_values=['Male'])
```
